```python
!jupyter nbextension enable --py widgetsnbextension
!pwd flights && rm flights.db
import sqlite3

conn = sqlite3.connect('flights.db')

c = conn.cursor()


# Create table
c.execute('''CREATE TABLE IF NOT EXISTS location(
				loc_id TEXT PRIMARY KEY,
				city TEXT NOT NULL,
				state TEXT NOT NULL,
				country TEXT NOT NULL,
				zip_code TEXT NOT NULL)''')

c.execute('''CREATE TABLE IF NOT EXISTS airline(
				airline_id TEXT PRIMAY KEY,
				airline_name INTEGER NOT NULL,
				is_charter INTEGER NOT NULL)''')

c.execute('''CREATE TABLE IF NOT EXISTS flight(
				flight_id TEXT PRIMAY KEY,
				seats INTEGER NOT NULL,
				airline_id TEXT NOT NULL,
				FOREIGN KEY(airline_id) REFERENCES airline(airline_id))''')
				
c.execute('''CREATE TABLE IF NOT EXISTS airport(
				airport_code TEXT PRIMARY KEY,
				airport_name TEXT NOT NULL UNIQUE,
				loc_id TEXT NOT NULL,
				FOREIGN KEY(loc_id) REFERENCES location(loc_id))''')

c.execute('''CREATE TABLE IF NOT EXISTS schedule(
				schedule_id TEXT PRIMARY KEY,
				flight_id TEXT NOT NULL,
				src_air_code TEXT NOT NULL,
				des_air_code TEXT NOT NULL,
				fare INTEGER NOT NULL,
				start_time DATETIME NOT NULL,
				end_time DATETIME NOT NULL,
				FOREIGN KEY(flight_id) REFERENCES flight(flight_id),
				FOREIGN KEY(src_air_code) REFERENCES airport(airport_code),
				FOREIGN KEY(des_air_code) REFERENCES airport(airport_code))''')

c.execute('''CREATE TABLE IF NOT EXISTS customer(
				customer_id INTEGER PRIMARY KEY,
				first_name TEXT NOT NULL,
				mid_name TEXT,
				last_name TEXT Not NULL,
				age INTEGER NOT NULL,
				address TEXT NOT NULL,
				loc_id TEXT NOT NULL,
				gender TEXT NOT NULL,
				email TEXT NOT NULL,
				phone TEXT NOT NULL,
				FOREIGN KEY(loc_id) REFERENCES location(loc_id))''')

c.execute('''CREATE TABLE IF NOT EXISTS booking(
				booking_id TEXT PRIMARY KEY,
				schedule_id TEXT NOT NULL,
				customer_id TEXT NOT NULL,
				price INTEGER NOT NULL,
				start_loc_id TEXT NOT NULL,
				end_loc_id TEXT NOT NULL,
				FOREIGN KEY(start_loc_id) REFERENCES location(loc_id),
				FOREIGN KEY(end_loc_id) REFERENCES location(loc_id),
				FOREIGN KEY(customer_id) REFERENCES customer(customer_id),
				FOREIGN KEY(schedule_id) REFERENCES schedule(schedule_id))''')

c.execute('''INSERT INTO location VALUES
                ('MUM','Mumbai','Maharashtra','India','400042'),
                ('DEL','New Delhi','Delhi','India','110057'),
                ('BEN','Bengaluru','Maharashtra','India','100015')''')

c.execute('''INSERT INTO airport VALUES
            ('BOM','Mumbai','MUM'),
            ('DEL','New Delhi','DEL'),
            ('BAN','Bengaluru','BEN')
        ''')
c.execute('''INSERT INTO airline values
                ('IGO','Indigo',0),
                ('AI','Air India',0),
                ('KF','King Fisher',1)''')

c.execute('''INSERT INTO customer values
                ( 1, 'Quinn', 'Sallie', 'Walter', 34, '225 Rutland Road', 'BOM', 'female', 'salliewalter@zensus.com', '+29 (851) 418-2048'),
                ( 2, 'Lucille', 'Roslyn', 'Cooley', 23, '856 Howard Alley', 'BEN', 'female', 'roslyncooley@zensus.com', '+51 (816) 425-3461'),
                ( 3, 'Julie', 'Cox', 'Cervantes', 27, '682 Montgomery Street', 'DEL', 'male', 'coxcervantes@zensus.com', '+33 (977) 577-2468'),
                ( 4, 'Orr', 'Lewis', 'Gillespie', 29, '300 Manhattan Court', 'BEN', 'male', 'lewisgillespie@zensus.com', '+14 (922) 575-2938'),
                ( 5, 'Marie', 'Hobbs', 'Walker', 31, '978 Herkimer Street', 'BOM', 'male', 'hobbswalker@zensus.com', '+20 (831) 401-2760'),
                ( 6, 'Jennifer', 'Hillary', 'Bridges', 40, '946 McKibben Street', 'BOM', 'female', 'hillarybridges@zensus.com', '+44 (974) 585-3937')
''')
# c.execute("INSERT INTO stocks VALUES ('2006-01-05','BUY','RHAT',100,35.14)")

# Save (commit) the changes
conn.commit()

# We can also close the connection if we are done with it.
# Just be sure any changes have been committed or they will be lost.
print("Done")
```
